```python
# I create a new data set containing numeric columns 
# then I will use the description to get the mean, the standart deviation and the median 
Num_columns = data._get_numeric_data() # get the numericall columns, incuding int and float
for col in Num_columns.columns:                      
    _, mean, std,_,_,median, *_ = Num_columns[col]                          # unpacking these values from descrition method 
    print(f"Column \033[1;32m{col.upper()} Grass\033[1;m")               # coloring the column's name 
    print(f"Mean : {mean}\nStandart Deviation :{std}\nMedian :{median}") 
```

```python
# I will fill the missing values with the mean(another way will be to fill in with median )
for col in Num_columns.columns:
    null = Num_columns[col].isnull().sum()
    if null:                                                          # if there missing values 
        Num_columns[col].fillna(Num_columns[col].mean(), inplace=True)   # replacing missing value by the mean 
    _, mean, std,_,_,median, *_ = Num_columns[col]
    print(f"Column \033[1;32m{col.upper()} Grass\033[1;m")
    print(f"Mean : {mean}\nStandart Deviation :{std}\nMedian :{median}")
```

```python
def LEARN_SIMPLE_LINREG(x, y):
    
    """this function return b0, b1 the parameters of the 
       the line Y = b1 * X + b0 """

    if x.shape != y.shape:
        raise RuntimeError('number of points do not match')
        
    x_m, y_m = x.mean(),y.mean() # we determine the mean of the the vector
    
    b1 = sum((x-x_m)*(y-y_m))/sum((x-x_m)**2)  # I use the formula from the slide 
    b0 = y_m - b1*x_m
    return b1, b0

def PREDICT_SIMPLE_LINREG(b1, b0, val):
    return b1*val + b0
```

```python
X = data['city-mpg'].values     # independent values 
Y = data['engine-size'].values # dependent values 
b1, b0 =  LEARN_SIMPLE_LINREG(X, Y)
Y_hat = PREDICT_SIMPLE_LINREG(b1, b0, X)
plt.subplots(figsize=(11, 9)) 
plt.scatter(X, Y, marker='o')
plt.plot(X, Y_hat, 'r')
```

WE can see that the model underfits, this is  due to the non linear behavours of our data. For example when we look at the left and right hand side of the of plot, the line does not fit to these points.
